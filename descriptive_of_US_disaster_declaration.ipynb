```python


# Create a pivot table
pt = combined_data.pivot_table(values='disaster_number', index='state', columns='year').fillna(0)

# Set the size of the heatmap figure
plt.figure(figsize=(12, 8))

# Generate the heatmap
sns.heatmap(pt, cmap='coolwarm')

# Save the heatmap as an image
plt.savefig('heatmap.png')

# Display the heatmap
plt.show()
```

The heatmap that results shows how frequently disasters occur in certain states over a range of years. Higher values are represented by warmer hues, which indicate that disasters are more likely to occur in certain states and years. Lower values are represented by cooler hues, which denote a lower frequency.
